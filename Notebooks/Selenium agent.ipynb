```python
from transformers import AutoModelForSeq2SeqLM
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("LucasThil/T5_large_hierarchy8_256_512")
model = AutoModelForSeq2SeqLM.from_pretrained("LucasThil/T5_large_hierarchy8_256_512")
```

```python
def make_inference(text):
  print("1 tokenizing inputs")
  inputs = tokenizer(text, return_tensors="pt").input_ids
  print("2 Generating model outputs")
  outputs = model.generate(inputs, max_new_tokens=100, do_sample=False)
  print("Get tokenizer prediction")
  prediction = tokenizer.decode(outputs[0], skip_special_tokens=True)
  print("Return prediction")
  return prediction
```

```python
import os
import torch

model_path = "/content/drive/MyDrive/webn_t5_3b"

state_dict = {}
for subdir in os.listdir(model_path):
    subdir_path = os.path.join(model_path, subdir)
    if os.path.isdir(subdir_path):
        for filename in os.listdir(subdir_path):
            if filename.startswith("state.param_states.decoder.layers_"):
                layer_name = filename.split(".")[1]
                parameter_name = filename.split(".")[2:]
                parameter_name[-1] = parameter_name[-1].split("_")[0]
                parameter_name = ".".join(parameter_name)
                if layer_name not in state_dict:
                    state_dict[layer_name] = {}
                state_dict[layer_name][parameter_name] = torch.load(os.path.join(subdir_path, filename))
            elif filename.startswith("target.decoder.layers_"):
                layer_name = filename.split(".")[1]
                parameter_name = filename.split(".")[2:]
                parameter_name[-1] = parameter_name[-1].split("_")[0]
                parameter_name = ".".join(parameter_name)
                if layer_name not in state_dict:
                    state_dict[layer_name] = {}
                state_dict[layer_name][parameter_name] = torch.load(os.path.join(subdir_path, filename))

# Create an instance of the T5 model class and load the state dict
model = T5ForConditionalGeneration(config=config)
model.load_state_dict(state_dict)
```
