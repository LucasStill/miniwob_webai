```python
from datasets import load_dataset

# Load dataset regarding action-step sequences
#raw_datasets_actions = load_dataset("LucasThil/miniwob_plusplus_hierarchical_training_actions")
raw_datasets_actions = load_dataset("LucasThil/miniwob_plusplus_hierarchical_training_actions_drain")

# Load dataset regarding task of hierarchical planning
raw_datasets_planning = load_dataset("LucasThil/miniwob_plusplus_hierarchical_planning")
```

```python
from os import RWF_NOWAIT
i = 0
# Removing click where action is type
history_episodes, instruction, actions, refs, keydown_text, subtask_completion = [], [], [], [], [], []

for row in raw_datasets_actions['train']:
  if 'Enter' in row['instruction'] or 'type' in row['instruction']:
    if 'click' in row['actions']:
        continue
    else:
        history_episodes.append(row['history_episodes'])
        instruction.append(row['instruction'])
        actions.append(row['actions'])
        refs.append(row['refs'])
        keydown_text.append(row['keydown_text'])
        subtask_completion.append(row['subtask_completion'])
  else:
      history_episodes.append(row['history_episodes'])
      instruction.append(row['instruction'])
      actions.append(row['actions'])
      refs.append(row['refs'])
      keydown_text.append(row['keydown_text'])
      subtask_completion.append(row['subtask_completion'])

  i += 1
  if i > 10 and False:
    break
```

```python
import pandas as pd
```

```python
data = {
    'history_episodes':history_episodes,
    'instruction': instruction,
    'actions': actions,
    'refs': refs,
    'keydown_text': keydown_text,
    'subtask_completion': subtask_completion
}
dataframe = pd.DataFrame(data)
```
