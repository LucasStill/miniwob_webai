```python
from transformers import AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")
```

```python
encoded_input = tokenizer("Do not meddle in the affairs of wizards, for they are subtle and quick to anger.")
print(encoded_input)
```

```python
tokenizer.decode(encoded_input['input_ids'])
```

```python
# Pass several sentences into the encoder as a list
batch_sentences = [
    "But what about second breakfast?",
    "Don't think he knows about second breakfast, Pip.",
    "What about elevensies?",
]
encoded_inputs = tokenizer(batch_sentences)
print(encoded_inputs)
```

# Padding:
Need to have same length, padding is needed. Add padding token to shorter sentences.

```python
encoded_input = tokenizer(batch_sentences, padding=True)
print(encoded_input)
```

# Truncation:
Sometimes too long so need to truncate

```python
encoded_input = tokenizer(batch_sentences, padding=True, truncation=True)
print(encoded_input)
```

# Build tensors
want to return actual tensors fed into the model

```python
encoded_input = tokenizer(batch_sentences, padding=True, truncation=True, return_tensors="pt")
print(encoded_input)
```

# Fine tune

```python
from datasets import load_dataset

dataset = load_dataset("yelp_review_full")
dataset['train'][100]
```

```python
def tokenize_function(examples):
  return tokenizer(examples["text"], padding='max_length', truncation=True)

tokenized_datasets = dataset.map(tokenize_function, batched=True)
```

```python
small_train_dataset = tokenized_datasets['train'].shuffle(seed=42).select(range(100))
small_eval_dataset = tokenized_datasets['test'].shuffle(seed=42).select(range(100))
```

```python
from transformers import AutoModelForSequenceClassification
#model=AutoModelForSequenceClassification.from_pretrained('/content/drive/MyDrive/WebAI/Notebooks/test_trainer', local_files_only=True)
model = AutoModelForSequenceClassification.from_pretrained('bert-base-cased', num_labels=5)
```

# Training hyper params

```python
import numpy as np
import evaluate
```

```python
metric = evaluate.load('accuracy')
```

```python
def compute_metrics(eval_pred):
  logits, labels = eval_pred
  predictions = np.argmax(logits, axis=-1)
  return metric.compute(predictions=predictions, references=labels)
```

```python
from transformers import TrainingArguments, Trainer
training_args = TrainingArguments(
    #output_dir="/content/drive/MyDrive/WebAI/Notebooks/test_trainer", 
    output_dir="/test_save_path",
    evaluation_strategy="epoch",
    overwrite_output_dir=True,
    num_train_epochs=1,
    logging_steps=1000,
    save_steps=1000,
    resume_from_checkpoint=True,
    push_to_hub=True,
    )
```

```python
trainer = Trainer(
    model=model,
    #model=AutoModelForSequenceClassification.from_pretrained('/content/drive/MyDrive/WebAI/Notebooks/test_trainer'),
    args=training_args,
    train_dataset=small_train_dataset,
    eval_dataset=small_eval_dataset,
    compute_metrics=compute_metrics,
)
```

```python
#trainer.train(resume_from_checkpoint=True)
#trainer.train(resume_from_checkpoint='/content/drive/MyDrive/WebAI/Notebooks/test_trainer')
trainer.train()
```
